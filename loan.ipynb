```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the categorical column
ds['Loan_ID'] = label_encoder.fit_transform(ds['Loan_ID'])
ds['Gender'] = label_encoder.fit_transform(ds['Gender'])
ds['Education'] = label_encoder.fit_transform(ds['Education'])
ds['Self_Employed'] = label_encoder.fit_transform(ds['Self_Employed'])
ds['Property_Area'] = label_encoder.fit_transform(ds['Property_Area'])
ds['Married'] = label_encoder.fit_transform(ds['Married'])
#accidentally removed the line for dependents column, it was also encoded
print(ds)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit and transform the categorical column
ds['Loan_Status'] = label_encoder.fit_transform(ds['Loan_Status'])

#print(ds)
```

```python
x = ds.drop(columns = 'Loan_Status', axis = 1)
y = ds['Loan_Status']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ds
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify = y, random_state = 1)
```

```python
model = svm.SVC(kernel = 'linear')
model.fit(x_train, y_train)
```

```python
train_pred = model.predict(x_train)
train_acc = accuracy_score(train_pred, y_train)
print('Accuracy is : ', train_acc)
```

```python
test_pred = model.predict(x_test)
test_acc = accuracy_score(test_pred, y_test)
print('Accuracy is : ', test_acc)
```
